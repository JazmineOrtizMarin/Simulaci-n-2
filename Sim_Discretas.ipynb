```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
import random as rd
import math
```

```python
mu = 4
k = 12
```

```python
p = poisson.pmf (k, mu)
print ("P(k=12,mu=4) = ", p)
```

```python
ks = range(0, 21) # Para la gr√°fica
pmf = poisson.pmf (ks, mu)
```

```python
plt.figure()
plt.bar(ks, pmf)
plt.scatter([k],[p], s=80, zorder=3)
plt.title(f"Poisson(mu={mu})-P(k={k}) = {p:.10f})")
plt.xlabel("k")
plt.ylabel("P(K=k)")
plt.show()
```

```python
# Par√°metros
lmbd = 4
T = 5
```

```python
# Inicializaci√≥n
t = 0
N = 0
S = [] # Tiempos de arribo
```

```python
while True:
  U = rd.random()
  t = t - (math.log(U) / lmbd)
  if t > T:
    break
  N += 1
  S.append (t)
```

```python
print("N√∫mero de llegadas N(T) = ", N)
print("N√∫mero de llegada S = ", S)
```

```python
plt.step([0] + S + [T], list(range(0,N+1))+[N], where='post')
plt.title(f"Proceso de Poisson Homog√©neo (Œª={lmbd},T={T})")
plt.xlabel("Tiempo")
plt.ylabel("N√∫mero de llegadas N(t)")
#plt.grid(True)
plt.show()
```
