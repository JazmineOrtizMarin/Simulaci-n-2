# **Representaci√≥n en Series de Fourier**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Valores del Ejercicio
Ms = 1, 2, 5, 10, 50, 200

# Dominio
t = np.linspace(0,2*np.pi,2000)
```

```python
def W_M(t, M):
    Z = np.random.normal(0, 1, M+1) # Z0 hasta ZM
    Wt = Z[0] * (t / np.sqrt(2*np.pi))

    # Suma de Fourier
    for n in range (1, M+1):
        Wt += Z[n] * np.sin(n*t/2) / n
    return Wt
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12,10))
axes = axes.flatten()

for idx, M in enumerate (Ms):
    Wt = W_M(t, M)
    axes[idx].plot(t, Wt)
    axes[idx].set_title(f'M = {M}')
    axes[idx].set_xlim([0, 2*np.pi])

plt.tight_layout()
plt.show()
```
