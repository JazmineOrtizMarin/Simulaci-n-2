```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
def g(x):
  return np.sqrt(np.arctan(x))
```

```python
X=np.linspace(0,1,100)
plt.plot(X,g(X))
```

```python
def crudo(N):
  G=[]
  for i in range (N):
    u=rd.random()
    G.append(g(u))
  return np.mean(G), np.var(G), np.std(G), G
```

```python
e=crudo(100)
```
