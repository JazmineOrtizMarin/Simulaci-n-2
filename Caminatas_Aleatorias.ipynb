```python
import random as rd
import matplotlib.pyplot as plt
```

```python
pasos = 2000
```

```python
def caminata_1D (U, D, pasos):
  x = 0
  posiciones = [x]

  for i in range(pasos):
    paso = rd.choice([U, D])
    x += paso
    posiciones.append(x)

  return posiciones
```

```python
U = 1
D = -1
```

```python
trayectoria = caminata_1D(U, D, pasos) # Ejecutamos 1D
```

```python
plt.plot(trayectoria)
plt.title("Caminata Aleatoria 1D")
plt.xlabel("Paso")
plt.ylabel("Posici√≥n")
plt.grid()
plt.show()
```

```python
def caminata_2D (pasos):
  x = 0
  y = 0
  posiciones_x = [x]
  posiciones_y = [y]

  for i in range(pasos):
    movimiento = rd.choice(['U', 'D', 'L', 'R'])

    if movimiento == 'U':
      y += 1
    elif movimiento == 'D':
      y -= 1
    elif movimiento == 'L':
      x -= 1
    elif movimiento == 'R':
      x += 1

    posiciones_x.append(x)
    posiciones_y.append(y)

  return posiciones_x, posiciones_y
```

```python
x_vals, y_vals = caminata_2D(pasos) # Ejecutamos 2D
```

```python
plt.plot(x_vals, y_vals, lw=1)
plt.scatter(x_vals[0], y_vals[0], color='green', label='inicio')
plt.scatter(x_vals[-1], y_vals[-1], color='red', label='fin')
plt.legend()
plt.title("Caminata Aleatoria 2D")
plt.xlabel("Posici√≥n en X")
plt.ylabel("Posici√≥n en Y")
plt.grid()
plt.axis()
plt.show()
```
